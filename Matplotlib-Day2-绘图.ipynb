# 第0节：Matplotlib绘图函数
Matplotlib提供了16种不同图象的绘制函数

以下提供一些例子

---

# 第1节：饼图
函数：plt.pie(data, explode)

```python
import matplotlib.pyplot as plt

labels = 'A', 'B', 'C', 'D'  # 标签
sizes = [15, 30, 45, 10]  # 每一个区域的比例
explode = (0, 0.1, 0, 0)  # 每一块是否突出来

plt.pie(sizes, explode=explode, labels=labels,
        autopct='%1.1f%%', shadow=False, startangle=90)
# autopct:百分数的呈现形式
# shadow:在饼图下面画一个阴影。默认值：False，即不画阴影
# startangle:起始绘制角度,默认是从x轴正方向逆时针画起,如设定90则从y轴正方向画起

plt.axis('equal')  # 该行代码使饼图长宽相等
plt.show()
```

# 第2节：直方图
---

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
mu, sigma = 100, 20
a = np.random.normal(mu, sigma, size=100)#生成一组服从正态分布的随机数

#bin：总共有几条条状图
plt.hist(a, 20, density=True, histtype='stepfilled', facecolor='b', alpha=0.75)
plt.title('Histogram')

plt.show()
```

# 第3节：极坐标图（了解）

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10
theta = np.linspace(0, 2*np.pi, N, endpoint=False)
radii = 10*np.random.rand(N)
width = np.pi/2*np.random.rand(N)

ax = plt.subplot(111, projection='polar')
bars = ax.bar(theta, radii, width=width, bottom=0.0)

for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.viridis(r/10.))
    bar.set_alpha(0.5)

plt.show()
```

# 第4节：散点图

```python
#PPT的例子不好，以下是更常用的例子
import matplotlib.pyplot as plt
import numpy as np


x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y, marker='o', color='c', s=40, label='Point')

plt.legend(loc='best')    # 设置 图例所在的位置 使用推荐位置

plt.show()
```
