# 第0节：Matplotlib简介

Matplotlib库由各种可视化类构成，内部结构复杂，受Matlab启发

matplotlib.pyplot是绘制各类可视化图形的命令子库，相当于快捷方式

----

```python
import matplotlib.pyplot as plt

# plt.plot()只有一个输入列表或数组时，参数被当作Y轴，X轴以索引自动生成
plt.plot([3,1,4,5,2])

plt.ylabel('Grade')

# plt.savefig()将输出图形存储为文件，默认PNG格式，可以通过dpi修改输出质量
plt.savefig('example.pdf',dpi = 1000)
plt.show()
```

```python
# plt.plot(x,y)当有两个以上参数时，按照X轴和Y轴顺序绘制数据点
x = [0,2,4,6,8]
y = [3,1,4,5,2]
plt.ylabel("Grade")

# plt.axis(x1,x2,y1,y2)
# x轴范围[x1,x2]
# y轴范围[y1,y2]
plt.axis([-1,10,0,6])

plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

def g(t):
    return np.cos(2*np.pi*t)

a = np.arange(0,5,0.02)

plt.subplot(2,1,1) # 等价于plt.subplot(211)
plt.plot(a,f(a))

plt.subplot(2,1,2)
plt.plot(a,g(a),'r--')

plt.show()
```

# 第1节：plot函数
plt.plot(x, y, format_string, $**kwargs$)

+ x : X轴数据，列表或数组，可选
+ y : Y轴数据，列表或数组
+ format_string: 控制曲线的格式字符串，可省略
+ $**kwargs$ : 第二组或更多(x,y,format_string)

+ **当绘制多条曲线时，各条曲线的x不能省略**

```python
import numpy as np
import matplotlib.pyplot as plt

# 正态分布
def f(x, mu, sig):
    return 1/(np.sqrt(2*np.pi)*sig)*np.exp(-1*(x-mu)**2/2*sig**2)

x = np.arange(-5, 5, 0.01)

plt.plot(x, f(x, 0, 1), x, f(x, 0, 0.5))
# 曲线1：x = x，y = f(x,0,1)
# 曲线2：x = x, y = f(x,0,0.5)

plt.show()
```

# 第2节：plot的中文显示

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 最简单的方法
plt.rcParams['font.family']=['SimHei'] #显示中文标签
plt.rcParams['axes.unicode_minus'] = False

# 正态分布
def f(x, mu, sig):
    return 1/(np.sqrt(2*np.pi)*sig)*np.exp(-1*(x-mu)**2/2*sig**2)

x = np.arange(-5, 5, 0.3)

plt.xlabel('x轴')
plt.ylabel('y轴')

plt.plot(x, f(x, 0, 1), 'c--')
plt.show()
```

### rcParams的属性
+ 'font.family' 用于显示字体的名字
+ 'font.style' 字体风格，正常'normal'或 斜体'italic'
+ 'font.size' 字体大小，整数字号或者'large'、'x‐small'

---

**字体**

+ 'SimHei' 中文黑体
+ 'Kaiti' 中文楷体
+ 'LiSu' 中文隶书
+ 'FangSong' 中文仿宋
+ 'YouYuan' 中文幼圆
+ 'STSong' 华文宋体

```python
# 方法2：在有中文输出的地方，增加一个属性：fontproperties
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
# 正态分布


def f(x, mu, sig):
    return 1/(np.sqrt(2*np.pi)*sig)*np.exp(-1*(x-mu)**2/2*sig**2)


x = np.arange(-5, 5, 0.3)

plt.xlabel('x轴',fontproperties='SimHei',fontsize = 15)
plt.ylabel('y轴',fontproperties='SimHei',fontsize = 15)

plt.plot(x, f(x, 0, 1), 'c--')
plt.show()
```

# 第3节：plot文本显示


```python
# plt.xlabel() 对X轴增加文本标签
# plt.ylabel() 对Y轴增加文本标签

# plt.title() 对图形整体增加文本标签

# plt.text(x,y,words) 在(x,y)处增加文本words

# plt.annotate() 在图形中增加带箭头的注解
# plt.annotate(s, xy=arrow_crd, xytext=text_crd, arrowprops=dict)
# xy：箭头位置
# xytext：文本位置
# arrowprops：样式


import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False

# 正态分布
def f(x, mu, sig):
    return 1/(np.sqrt(2*np.pi)*sig)*np.exp(-1*(x-mu)**2/2*sig**2)


x = np.arange(-5, 5, 0.3)

plt.title('正态分布曲线')
plt.xlabel('x轴')
plt.ylabel('y轴')

plt.text(2, 0.3, 'u = 0,sigma = 1')
plt.annotate('标准正态分布', xy=(1, 0.3), xytext=(1.5, 0.4),
             arrowprops=dict(facecolor='black', shrink=0.1, width=1))

plt.grid(True)  # 是否设置网格线
plt.plot(x, f(x, 0, 1), 'r--')
plt.show()
```

# 第4节：子绘图区域（了解）

```python
# subplot2grid(shape, loc, rowspan=1, colspan=1, fig=None, **kwargs)

# shape：放置子图的网格规格，包含两个整型的元组，第一个代表网格的行数，第二个代表网格的列数
# loc：子图所在的位置，包含两个整型的元组，第一个代表子图所在的行数，第二个代表子图所在的列数

ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)  
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)  #col 显示图形占2列
ax3 = plt.subplot2grid((3,3), (1, 2), rowspan=2)  #row 显示图形占2行
ax4 = plt.subplot2grid((3,3), (2, 0))  
ax5 = plt.subplot2grid((3,3), (2, 1))  
plt.suptitle("subplot2grid")
plt.show()
```
